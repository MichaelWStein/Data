```python
import pandas as pd

prov_lan_df = pd.read_pickle("prov_languages.pkl")


prov_list = prov_lan_df.loc[0]

print(prov_list[0])

seats = ["seats", None, 338, 7, 4, 11, 10, 78, 121, 14, 14, 34, 42, 1, 1, 1]

prov_lan_df
```

```python
lan_df = prov_lan_df.append(pd.Series(seats, index=prov_lan_df.columns ), ignore_index=True)

lan_df
```

```python

calc = []
ref_str = ['votes', 'votes per seat']
y=0

for column in lan_df:
    # print(column)
    # print( type(lan_df.loc[2, column]))
    if isinstance(lan_df.loc[2, column], float):
        ref = lan_df.loc[2, column]/lan_df.loc[21, column]
    else:
        ref = ref_str[y]
        y=y+1
        
    calc.append(ref)
    # print(calc)
    #== 'class float'):
    #    print("float")
    #= 'float':
    #        ref = lan_df.loc[2, column]/lan_df.loc[21, column]
    #print(ref)

#type([ref])
lan_df = lan_df.append(pd.Series(calc, index=lan_df.columns ), ignore_index=True)

lan_df
```
